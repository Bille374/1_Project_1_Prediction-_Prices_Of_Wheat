# Urea Monthly Price - US Dollars per Metric Ton

```python
US_Dollar_metric_ton = pd.read_html("https://www.indexmundi.com/commodities/?commodity=urea&months=360")[1]
```

```python
US_Dollar_metric_ton
```

```python
US_Dollar_metric_ton_month = []
US_Dollar_metric_ton_year = []
for i, r in US_Dollar_metric_ton.iterrows():
    US_Dollar_metric_ton_month.append(r["Month"].split()[0])
    US_Dollar_metric_ton_year.append(r["Month"].split()[1])
    
```

```python
US_Dollar_metric_ton["month_name"] = US_Dollar_metric_ton_month
```

```python
US_Dollar_metric_ton["year"] = US_Dollar_metric_ton_year
```

```python
dates = pd.to_datetime(US_Dollar_metric_ton["Month"])
```

```python
US_Dollar_metric_ton["month_name2"] = dates.dt.month_name()
```

```python
US_Dollar_metric_ton
```

```python
US_Dollar_metric_ton.to_csv("US_Dollar_metric_ton-Update")
```
